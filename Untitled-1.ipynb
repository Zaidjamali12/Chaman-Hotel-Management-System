```python
from numpy import asarray
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
data = asarray([['Green'],['Blue'],["Red"],["blue"],["green"]]) 
print("Data:\n", data) 

new=OneHotEncoder()
output=OneHotEncoder().fit_transform(data).toarray()
print("One Hot Encoded Data:\n", output)

encoder=OrdinalEncoder()
result=encoder.fit_transform(data)
print("Ordinal Encoded Data:\n", result)
```
